To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2:1b'
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")
openAiClient = OpenAI(api_key=api_key)
ollamaClient = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
```

```python
# here is the question; type over this to ask something new
instruction = """
you are a python programming language expert who is helping an experienced frontend react developer understand python.
You first compare the code syntax and methods to javascript, then break down the code and explain what the code does.
"""
question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""

messages = [
    {"role": "system", "content": instruction},
    {"role": "user", "content": question}
]
```

```python
# Get gpt-4o-mini to answer, with streaming

stream = openAiClient.chat.completions.create(
    model=MODEL_GPT,
    messages=messages,
    stream=True
)
for chunk in stream:
    print(chunk.choices[0].delta.content, end="")
```

```python
# Get Llama 3.2 to answer
llamaStream = ollamaClient.chat.completions.create(
    messages=messages,
    model=MODEL_LLAMA,
    stream=True
    )

for chunk in llamaStream:
    print(chunk.choices[0].delta.content, end="")
```
